```python
# Linear Regression

# data no behavour linear hovo joiye to j use thay(y=mx+c)

# output only input per depend hoy tene linear regresition kevay
# output ek j parameter per depend hoy to tene simple linear regration kevay
# je pridict karvani hoy tene 'y' ma store karvu
# y  e dependent hoy and x e indipendent hoy
# m=co-efficient
# c=intercept


```

```python
import pandas as pd
data=pd.read_csv("book1.csv")
df=pd.DataFrame(data)
df
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(df["cgpa"],df["package"])
plt.xlabel("CGPA")
plt.ylabel("PACKAGE")
plt.show()
```

```python
y=df["package"]
y.head()
```

```python
x=df[["cgpa"]]
x.head()
# (x e 2D hoy)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)#2d
print(x_test.shape)#2d
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
m=lr.coef_
m
```

```python
c=lr.intercept_
c
```

```python
y=m*(5.65)+c
y
```

```python
y2=lr.predict([[5.65]])
y2
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
# tetsing 
from sklearn.metrics import mean_squared_error
print("MSE",mean_squared_error(y_test,y_pred))
```
