# Support Vector Machine(SVM)



```python
import pandas as pd
df=pd.read_csv("Employee_Attrition.csv")# attrition =emplloyee job ma rehshe ke nahi
df.isna().sum()
df.head()
```

```python
df.drop(columns="Employee_ID",inplace=True)

df=pd.get_dummies(data=df,drop_first=True)
df.head()
```

```python
df.columns
x=df[['Age', 'Monthly_Income', 'Years_at_Company', 'Job_Satisfaction',
       'Distance_From_Home','Job_Role_Developer', 'Job_Role_HR',
       'Job_Role_Manager', 'Work_Life_Balance_Good', 'Work_Life_Balance_Poor',
       'Overtime_Yes', 'Education_Postgraduate', 'Education_School']]
y=df["Attrition"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC
model=SVC(C=10,random_state=42)#c=margin
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
tn=cm[0][0]
fp=cm[0][1]
fn=cm[1][0]
tp=cm[1][1]
print("tn ",tn)
print("fp ",fp)
print("fn ",fn)
print("tp ",tp)
```

```python
total=tn+tp+fn+fp
acc=(tn+tp)/total
error=(fn+fp)/total
sensi=tp/(tp+fn)
spe=tn/(tn+fp)
prec=tp/(tp+fp)
print("accuracy :",acc)
print("Error_rate :",error)
print("sensitivity :",sensi)
print("specifivity :",spe)
print("precison :",prec)
df.head(0)
```

```python
import pandas as pd
df=pd.read_csv("Employee_Attrition.csv")# attrition =emplloyee job ma rehshe ke nahi
df.isna().sum()
df.head()
```

```python
df.drop(columns="Employee_ID",inplace=True)

df=pd.get_dummies(data=df,drop_first=True)
df.head()
```

```python
df.columns
y=df["Attrition"]
x=df[['Age', 'Monthly_Income', 'Years_at_Company', 'Job_Satisfaction',
       'Distance_From_Home','Job_Role_Developer', 'Job_Role_HR',
       'Job_Role_Manager', 'Work_Life_Balance_Good', 'Work_Life_Balance_Poor',
       'Overtime_Yes', 'Education_Postgraduate', 'Education_School']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
accuracy=[]
kn=[]
for i in range(1,26):
    nn=KNeighborsClassifier(n_neighbors=i)
    nn.fit(x_train,y_train)
    y_pred=nn.predict(x_test)
    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)
    tn=cm[0][0]
    fp=cm[0][1]
    fn=cm[1][0]
    tp=cm[1][1]
    total=tn+tp+fn+fp
    acc=(tn+tp)/total
    print("accuracy :",acc)
    accuracy.append(acc)
    kn.append(i)
ACCURAXY=(max(accuracy))
maxK=accuracy.index(max(accuracy))
print(ACCURAXY)
print(maxK+1)
```

```python
nn=KNeighborsClassifier(n_neighbors=2)
nn.fit(x_train,y_train)
```

```python
y_pred=nn.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
